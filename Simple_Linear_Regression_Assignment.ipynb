```python
# import libraries what we need for this project.

import pandas as pd;                    # Analyze data(cleaning and plotting) / Mutable for two dimensational data structure.
import numpy as np;                     # Numpy provide an array object that faster than traditional python lists.
import matplotlib.pyplot as mat;        # Visualization for 2D plots of arrays like scatter, pie charts, graphs, line, bar, histograms and etc.
```

```python
# Data collection / Fetch data using from "pandas" library. 

dataset = pd.read_csv("Salary_Data.csv");       # Read "csv" file and returns in table format.

#print(dataset);           # if we use print statement showing normal behaviour of dataframe. 

dataset                    # if we execute without print statement showing advance behaviour of dataframe.
```

```python
# Input split

indep = dataset[["YearsExperience"]]; # fetch independent data from "dataset" / fetch only independent data from Salary_Data table.

#print(indep);

indep      # Execute and display(print) independent values.
```

```python
# Output split

dep = dataset[["Salary"]];  # fetch dependent data from dataset / fetch only dependent data from Salary_Data table. 

#print(dep);

dep                # Execute and display(print) dependent values.
```

```python
# Develop Scatter / Display graph using from "matplotlib.pyplot" library.

mat.scatter(indep, dep);             # put independent and dependent data as arguments to scatter function using "mat" library.

mat.xlabel("Years of Experience");   # display x axis name using xlabel function.

mat.ylabel("Salary");                # display y axis name using ylabel function.

#print(mat);                         

#mat

mat.show();                 # To show enhanced avoid unnessesary codes.
```

```python
# To split training and test set from original dataset

from sklearn.model_selection import train_test_split;   # import "train_test_split" function from "model_selection" class using "sklearn" library.

# put independent and dependent data, test size, random state as parameters to "train_test_split" function. 
# split the independent and dependent data to respective variables(nessesary variable counts - 4 to train(2) and test(2))
x_train, x_test, y_train, y_test = train_test_split(indep, dep, test_size = 0.30, random_state = 0);

```

```python
# Display x_train

#print(x_train);

x_train     # "x_train" variable shuffled for randam state wise. (random_state = 0)
```

```python
# Display y_train variable using different random state value. (randam_state = 5)

y_train
```

```python
# get rows and columns total counts / view rows and columns total counts from respective dataset

print(y_test.shape);  # y_test row and column total counts.

print(y_train.shape); # y_train row and column total counts.

print(x_test.shape);  # x_test row and column total counts.

print(x_train.shape); # x_train row and column total counts.

# get row counts only / view row count only

print("\n", len(x_train));
```

```python
# Model creation for simple linear regression

# import "LinearRegression" functioin from "linear_model" class using "sklearn" library.
# "LinearRegression" function have a model creation formula.
from sklearn.linear_model import LinearRegression; 

regressor = LinearRegression(); # assign a function to an object / # "LinearRegression" function to store a formula in "regressor" variable. 

# Model created using "fit" function
regressor.fit(x_train, y_train); # substituted "x_train" and "y_train" as arguments to "fit" function / only pass, training values to "fit" function. 

# weight and bias calulate
weight = regressor.coef_; # calculate weight(slope) using "coef_" function / weight generated and assigned to "weight" variable.

#print(weight); 

weight # here we calculate how salary increases in which ratio using weight(slope) value. / display or print weight value.
```

```python
bias = regressor.intercept_; # calculate bias using "intercept_" function / bias calculated and assigned to "bias" variable.

bias # here we calculate starting value where to start origin point of salary / like fresher salary calculated / display or print bias value.
```

```python
# Evaluation process  - here we calculate the model which we have found as good or poor.

y_pred = regressor.predict(x_test) # To predict test set values, substituted "x_test" as parameter to "predict" function.

y_pred
```

```python
# created dataset for actual value and predicted value

# creaed empty table to "predActual" variable using "DataFrame" function with defined index value.
predActual = pd.DataFrame(index = range(0, 9)); 

#print(predActual);

predActual  # Emply table created and display with index(serial) value.
```

```python
predActual["Actual Value"] = y_test; # assign test set values to "predActual["Actual Value"]" variable.

y_test.index = range(0, 9); # test set value assigned as index wise using index function to "y_test" given through range values.

# y_test 

predActual
```

```python

y_pred # y_pred has array format.

# given "y_pred" value and "columns" as parameter to "DataFrame" function using "pandas" library, to "y_pred_table" variable.
y_pred_table = pd.DataFrame(y_pred, columns = ["Predicted Value"]); # "y_pred" created array format to table format.

y_pred_table

```

```python
predActual["Predicted Value"] = y_pred_table; # assign "y_pred_table" value to "predActual[predicted Value]".

predActual # displayed actual value and predicted value.
```

```python
# Evaluation metrics - find "r2_score". / Evaluation metrics only for test set in AI.

# In test set, evaluation matrics only calculated to Actual value and predicted value.

from sklearn.metrics import r2_score; # import "r2_score" function from "metrics" class using "sklearn" library.

# "r2_score" function stroed in "r" variable.
r = r2_score(y_test, y_pred); # put test set values only("y_test", "y_pred") as arguments to "r2_score" function.

# if the value of "r" nearly to 1 then that is "Good performence" / our model is Good

# if the value of "r" nearly to 0 then that is "Poor performence"  / our model is poor

r          # display "r" value.
```

```python
# save the best model

import pickle; # import "pickle" library

filename = "finalModel_1.sav" # put "finalModel_1.sav" file name with ".sav" extension to "filename" variable 

# given "regressor" and "filename("wb" as write file)" as parameter to "dump" function.
pickle.dump(regressor, open(filename, "wb")); # ".sav" file stored in respective repository(python workstation(Jupyter notebook)).
```

```python
# Deployemnt - Load and deploy the model

# given "finalModel_1.sav(file name)" and "rb(read file)" as parameter to "load" function using pickle library
load_model = pickle.load(open("finalModel_1.sav", "rb")); # "load" function stored in "load_model" variable.

# using "predict()" function to calculate the expected result / To calculate the "predicted" value what we have expect.
result = load_model.predict([[15]]);  # given independent value what we expect stored in "result" variable.

result    # Expected and predicted value is displayed(salary for 15 years of experienced person).


```
